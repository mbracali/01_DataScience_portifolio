# Criando a matriz de confusão

Para criarmos a matriz de confusão precisamos dividir o dataframe de treino em 2 e compara-los para entender a acuracia do modelo

```python
# Aqui utilizamos uma funcao que separa o dataframe em teste e treino, onde X e a massa de dados e Y e a variavel target
X_train, X_test, y_train, y_test = train_test_split(df_treino_trat.drop(['TARGET'], axis = 1), df_treino_trat['TARGET'],test_size=0.3, random_state=42)
```

A próxima linha é a linha que efetivamente executa a regressão logistica. Os parametros escolhidos são padroes

```python
clf_mat = LogisticRegression(random_state=0, solver='lbfgs',multi_class='auto', max_iter =100000).fit(X_train, y_train)
```

```python
y_predictions = clf_mat.predict(X_test)
```

```python
conf_matriz = confusion_matrix(y_test, y_predictions)
print(conf_matriz)
```

```python
score_back_test = clf_mat.score(X_train, y_train)
print(score_back_test)
```

Através de uma analise de correlação de variaveis, podemos verificar que a seleção de variaveis faz sentido

```python
# Dados originais
corr = df_treino.corr()
fig = plt.figure(num=None, figsize=(15, 15), dpi=80, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(df_treino.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df_treino.columns)
ax.set_yticklabels(df_treino.columns)
plt.show()
```
