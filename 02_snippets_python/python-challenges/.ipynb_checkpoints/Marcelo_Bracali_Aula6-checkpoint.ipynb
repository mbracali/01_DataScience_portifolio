```python
import matplotlib.pyplot as plt
import pandas as pd
import altair as alt
from vega_datasets import data
dfPokemon = pd.read_csv('PokemonMBA.csv')
```

## 1) Utilizando o dataset Pokemon, crie uma função que receba o dataset e o atributo e retorne um histograma que mostre a distribuição do valor dos atributos “attack”, “defense”,”Sp.Atk”, “Sp.Def” e “Speed”. Insira uma linha indicando o valor médio do atributo

```python
def func_Histogram(df, at):
  plt.hist(df[at], bins='auto', color= 'Blue', alpha=0.7, rwidth=0.85)
  plt.xlabel('Value')
  plt.ylabel('Frequency')
  plt.title(f'Histogram {at}')
  plt.axvline(df[at].mean(), color= 'b', linestyle='dashed', linewidth=2)
```

```python
func_Histogram(dfPokemon, 'Attack')
```

```python
func_Histogram(dfPokemon, 'Defense')
```

```python
func_Histogram(dfPokemon, 'Speed')
```

## 2) Usando o dataset Pokemon, crie um scatter plot comparando ataque (eixo x) e defesa (eixo y) de pokemons dos tipos Fire e Water. 

```python
fire = dfPokemon[dfPokemon['Type 1'] == 'Fire']
water = dfPokemon[dfPokemon['Type 1'] == 'Water']

plt.scatter(fire['Attack'], fire['Defense'], label='Fire', color='red')
plt.scatter(water['Attack'], water['Defense'], label='Water', color='blue')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title("Pokémon: Fire and Water")
plt.legend()
plt.show()
```

## 3) Usando o dataset Pokemons, crie um gráfico de linha para comparar a quantidade de pokemons de cada tipo ao longo das gerações

```python
dfPokemon.groupby(['Generation','Type 1']).size().unstack().plot().legend(loc='best',bbox_to_anchor=(0.9, 0.5, 0.5, 0.5))
```

## 4) Utilizando o dataset airports (var = data.airports()):
    -> Faça um rank bar dos 5 primeiros estados dos EUA com maior número de aeroportos


```python
airports = data.airports()

fiveAirportsUSA = (airports[airports['country'] == 'USA'])


alt.Chart(fiveAirportsUSA).transform_aggregate(
    count= 'count()',
    groupby= ['state']
).transform_window(
    rank= 'rank(count)',
    sort= [alt.SortField('count', order='descending')]
).transform_filter(
    alt.datum.rank <= 5
).mark_bar().encode(
    y= alt.Y('state:N',
    sort= alt.EncodingSortField(field='count', op='sum', order='descending')),
    x= 'count:Q',
    color= alt.Color('state')
).properties(
    title='Rank - Top 5 USA airports'
)
```

## 5) Utilizando o conjunto de dados íris (var = data.íris()):
    -> Crie dois gráficos scatter, chart1 e chart2. Para o chart1, x=petalLenght, y=petalWidth. Para o chart2, x = sepalLenght, y=sepalWidth.

    -> Componha esses gráficos numa única apresentação. Para isso, faça chart1 | chart2


```python
iris = data.iris()

fig = plt.figure()
chart1 = fig.add_subplot(1,3,1)
chart2 = fig.add_subplot(1,3,3)

chart1.scatter(iris['petalLength'], iris['petalWidth'],color='b')
chart1.set_title('Petal')
chart1.set_xlabel('Length')
chart1.set_ylabel('Width')

chart2.scatter(iris['sepalLength'], iris['sepalWidth'], color='r')
chart2.set_title('Sepal')
chart2.set_xlabel('Length')
chart2.set_ylabel('Width')
```

## 6) Usando o dataset Pokemon, crie um rank bar que mostre o rank para Type 1

```python
alt.Chart(dfPokemon).transform_aggregate(
    count= 'count()',
    groupby= ['Type 1']).transform_window(
    rank= 'rank(count)',
    sort= [alt.SortField('count', order='descending')]).mark_bar().encode(
    y= alt.Y('Type 1:N',
    sort= alt.EncodingSortField(field='count', op='sum', order='descending')),
    x= 'count:Q',
    color= alt.Color('Type 1')).properties(
    title='Pokemon: Rank for Types'
)
```
